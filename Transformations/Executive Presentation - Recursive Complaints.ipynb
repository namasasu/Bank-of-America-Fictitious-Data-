## Root Cause Analysis: Why Do Complaints Recur?

### We tested 8 hypotheses across 62,516 complaints. Four factors emerged as primary drivers:

---

### 🎯 **ROOT CAUSE #1: Resolution Type (Strongest Factor)**

**Finding:** How we resolve complaints has the biggest impact on whether they recur.

| Resolution Type | Recurrence Rate | Avg Complaints per Pattern | Relief Provided |
|----------------|-----------------|---------------------------|------------------|
| Closed with explanation | **76.58%** | **29.21** | 16.32% |
| Closed with monetary relief | 66.67% | 11.53 | 78.25% |
| Closed with non-monetary relief | **54.72%** | **5.74** | 80.83% |

**Key Insight:** Providing relief (monetary or non-monetary) reduces recurrence by **40%** and complaints by **5x**.

**Business Implication:** Simply explaining an issue without fixing it creates a costly "explain and repeat" cycle. One $500 relief payment is cheaper than handling 29 repeat complaints at $50 each ($1,450).

---

### 🚨 **ROOT CAUSE #2: Systemic Product Issues**

**Finding:** Certain complaint issues have 100% recurrence because they're product/process failures, not customer issues.

| Issue Category | Recurrence | Avg Complaints | Total Volume |
|----------------|-----------|----------------|-------------|
| **Managing an account** | **100%** | **296.25** | **15,109 (24%)** |
| Problem with purchase on statement | 100% | 86.57 | 4,415 |
| Trouble during payment process | 100% | 56.54 | 2,827 |
| Opening an account | 100% | 55.61 | 2,725 |

**Key Insight:** "Managing an account" generates **296 average complaints per pattern** (11x overall average). This is a systemic product quality failure.

**Business Implication:** Complaint handlers cannot fix this. Requires cross-functional task force (Product + Engineering + Operations) to identify and fix upstream root causes.

---

### 📦 **ROOT CAUSE #3: Product Complexity**

**Finding:** Certain products generate consistently high recurrence due to inherent complexity and recurring customer touch points.

| Product | Recurrence Rate | Chronic Patterns | Total Complaints |
|---------|----------------|-----------------|------------------|
| Credit card or prepaid card | 83.04% | 389 | 16,197 |
| Mortgage | 80.78% | 154 | 6,601 |
| Checking or savings account | 80.25% | 207 | 24,814 |
| Debt collection | 78.88% | 105 | 2,736 |

**Key Insight:** These products have 2.9x higher recurrence than student loans (28%). Complexity drives recurring problems.

**Business Implication:** Requires product-specific quality improvement programs, not just better complaint handling.

---

### 📍 **ROOT CAUSE #4: Geographic Concentration**

**Finding:** Certain states show extreme recurrence patterns, especially California.

| State | Recurrence Rate | Avg Complaints per Pattern | Total Complaints |
|-------|----------------|---------------------------|------------------|
| **California** | **87.10%** | **147.41** | **13,709** |
| Maryland | 87.50% | 30.61 | 1,959 |
| Georgia | 86.67% | 38.95 | 2,921 |
| Florida | 84.52% | 77.24 | 6,488 |

**Key Insight:** California averages **147 complaints per pattern** (5.5x national average). Regional service quality or process differences.

**Business Implication:** Requires regional investigation and best-practice replication from low-recurrence states.

### Methodology Notes
**Customer Proxy Approach:**
Since the dataset lacks direct customer IDs, we created customer proxies using `state + product + issue` combinations. This identifies complaint "profiles" that likely represent same customer scenarios.

**Recurrence Categories:**
* One-Time: 1 complaint
* Repeat (2-4): 2-4 complaints
* Frequent (5-9): 5-9 complaints
* Chronic (10+): 10+ complaints

**Limitations:**
* Customer proxy may not perfectly represent individual customers
* Analysis assumes complaints in same state+product+issue pattern are related
* Cannot distinguish between same customer filing multiple times vs. different customers with same issue
* Despite limitations, patterns are statistically significant and actionable
